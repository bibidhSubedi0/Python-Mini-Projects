```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
```

```python
file_path = r"C:\Users\Bibidh Subedi\.cache\kagglehub\datasets\uciml\zoo-animal-classification\versions\1\zoo.csv"

# Load the dataset using pandas
df = pd.read_csv(file_path)
```

```python
df
```

```python
feat = df.iloc[:,1:-1]
targ = df.iloc[:,-1]
x = np.array(feat)
y=np.array(targ)
```

```python
import torch
from torch.utils.data import TensorDataset
```

```python
dataset = TensorDataset(torch.tensor(x),torch.tensor(y))
```

```python
from torch.utils.data import DataLoader
batch_size = 4
suffel = True

dl = DataLoader(dataset,batch_size, suffel)
```

```python
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
model = nn.Sequential(nn.Linear(16,32),nn.Linear(32,32),nn.Linear(32,8),nn.Softmax())

crt =nn.CrossEntropyLoss()
opt = optim.SGD(model.parameters(), lr = 0.001)
```

```python
for ep in range(1000):
    for data in dl:
        opt.zero_grad()
        feat, tar = data
        feat = feat.float()
        pred = model(feat)
        loss = crt(pred,tar)
        loss.backward()
        opt.step()
```

```python

data_iter = iter(dl) 
sample_data, sample_label = next(data_iter) 

sample_data = sample_data.float()

model.eval()  # Set the model to evaluation mode
with torch.no_grad():  # Disable gradient computation for inference
    outputs = model(sample_data) 

probs = torch.nn.functional.softmax(outputs, dim=1)  # Softmax to get probabilities for each class
_, predicted_classes = torch.max(probs, 1)  # Get the class index with the highest probability for each example


for i in range(len(predicted_classes)):
    print(f"Sample {i+1}:")
    print(f"Input: {sample_data[i]}")
    print(f"True label: {sample_label[i]}")
    print(f"Predicted label: {predicted_classes[i].item()}") 

    # Check if prediction is correct
    if predicted_classes[i].item() == sample_label[i].item():
        print("Correct!")
    else:
        print("Incorrect.")
    print("-" * 30)
```
